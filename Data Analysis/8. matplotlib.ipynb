- 엑셀 파일에서 병합된 셀을 데이터프레임에서 변환할 때 적절한 값을 찾지 못해 전출지별 열에 결측치 발생
    - 누락데이터 앞 행의 데이터로 채워야 함

```python
df = df.fillna(method = "ffill")  # 앞에 나온 데이터로 뒷부분 채워준다
```

```python
df.head()
```

```python
df.tail()
```

```python
# 서울에서 다른 지역으로 이동한 데이터만 추출하여 정리
df_seoul = df[(df["전출지별"] == "서울특별시") & (df["전입지별"] != "서울특별시")]
df_seoul = df_seoul.drop(["전출지별"], axis = 1)  # 어차피 전출지가 서울특별시만 있으니까 전입지만 보이게 끔
df_seoul.rename({"전입지별" : "전입지"}, axis = 1, inplace = True)  # 이름 전입지로 변경
df_seoul.set_index("전입지", inplace = True)
df_seoul.head()
```

```python
# 서울에서 경기도로 이동한 데이터만 선택
sr_gy = df_seoul.loc["경기도"]   # 컬럼명을 아니까 loc가 더 편리함
```

```python
sr_gy.head()
```

```python
# 시리즈의 인덱스를 x축, 값을 y축으로 선 그래프 그리기
plt.plot(sr_gy)
plt.title("서울 -> 경기 인구 이동")
plt.xlabel("기간")
plt.ylabel("이동 인구수")
plt.show()
```

## 그래프 꾸미기

- 눈금 레이블이 들어갈 충분한 여유공간이 없으면 글씨가 겹치는 문제가 발생
- 문제 해결을 위한 방법
    - figure() 로 공간을 만들기 위해 그래프의 가로 사이즈를 더 크게 설정
    - xticks() 로 x축 눈금 레이블을 90도 회전시켜서 글씨가 겹치지 않게 하기

```python
plt.figure(figsize = (14, 5))
plt.xticks(rotation = "vertical")   # x축 90도 회전, "vertical" 대신 숫자를 넣어도 괜찮음
# plt.plot(sr_gy.index, sr_gy.values)   = plt.plot(sr_gy) 같은거임
plt.plot(sr_gy)
plt.title("서울 -> 경기 인구 이동")
plt.xlabel("기간")
plt.ylabel("이동 인구수")

# 범례 표시
plt.legend(labels = ["서울 -> 경기"])
plt.show()
```

```python
# 스타일 서식 지정
plt.style.use("ggplot")    # 얜 한번만 지정해주면 계속 안 써줘도 됨

plt.figure(figsize = (14, 5))
plt.xticks(rotation = "vertical")   # x축 90도 회전, "vertical" 대신 숫자를 넣어도 괜찮음
# plt.plot(sr_gy.index, sr_gy.values)   = plt.plot(sr_gy) 같은거임
plt.plot(sr_gy, marker = "o", markersize = 10)  # 마커 표시 추가
plt.title("서울 -> 경기 인구 이동")
plt.xlabel("기간")
plt.ylabel("이동 인구수")

# 범례 표시
plt.legend(labels = ["서울 -> 경기"], fontsize = 15)
plt.show()
```

```python
# 그래프 객체 생성 (figure에 1개의 서브플롯을 생성)
fig = plt.figure(figsize = (20, 5))
ax = fig.add_subplot(1, 1, 1)

# axe 객체에 plot 함수를 그래프 출력
ax.plot(sr_gy, marker = "o", markersize = 10, markerfacecolor = "orange",
        color = "olive", linewidth = 2, label = "서울 -> 경기")
ax.legend()

# y축 범위 지정(최솟값, 최댓값)
ax.set_ylim(50000, 800000)

# 차트 제목 추가
ax.set_title("서울 -> 경기 인구 이동", size = 30)

# 축 이름 추가
ax.set_xlabel("기간", size = 12)
ax.set_ylabel("이동 인구 수", size = 12)

# 축 눈금 레이블 지정 및 75도 회전
ax.set_xticklabels(sr_gy.index, rotation = 75)

# 축 눈금 레이블 크기
ax.tick_params(axis = "both", labelsize = 10)

plt.show()
```

```python
# 서울에서 충청남도, 경상북도, 강원도로 이동한 인구 데이터 값만 선택
col_years = list(map(str, range(1970, 2018)))
df_3 = df_seoul.loc[["충청남도", "경상북도", "강원도"], col_years]
```

```python
df_3.head()
```

```python
# 그래프 객체 생성 (figure에 1개의 서브플롯을 생성)
fig = plt.figure(figsize = (20, 5))
ax = fig.add_subplot(1, 1, 1)

# axe 객체에 plot 함수를 그래프 출력
ax.plot(col_years, df_3.loc["충청남도"], marker = "o", markerfacecolor = "green",
        markersize = 10, color = "olive", linewidth = 2, label = "서울 -> 충남")
ax.plot(col_years, df_3.loc["경상북도"], marker = "o", markerfacecolor = "blue",
        markersize = 10, color = "skyblue", linewidth = 2, label = "서울 -> 경북")
ax.plot(col_years, df_3.loc["강원도"], marker = "o", markerfacecolor = "red",
        markersize = 10, color = "magenta", linewidth = 2, label = "서울 -> 강원")

# 범례 표시
ax.legend()

# 차트 제목 추가
ax.set_title("서울 -> 충남, 경북, 강원 인구 이동", size = 20)

# 축 이름 추가
ax.set_xlabel("기간", size = 12)
ax.set_ylabel("이동 인구 수", size = 12)

# 축 눈금 레이블 지정 및 75도 회전
ax.set_xticklabels(sr_gy.index, rotation = 75)

# 축 눈금 레이블 크기
ax.tick_params(axis = "both", labelsize = 10)

plt.show()
```

- 지리적으로 가까운 충남지역으로 이동한 인구가 다른 두 지역에 비해 많은 편
- 70 ~ 80 년대에는 서울에서 지방으로 전출하는 인구가 많았으나 90년 이후로는 감소하는 패턴을 보임

### 면적 그래프(area plot)

- 각 열의 데이터를 선 그래프로 구현하는데, 선 그래프와 x축 사이의 공간에 색이 입혀짐

```python
df_4 = df_seoul.loc[["충청남도", "경상북도", "강원도", "전라남도"], col_years]
df_4 = df_4.transpose()
df_4.head()
```

```python
# 면적 그래프 그리기
df_4.plot(kind = "area", stacked = False, alpha = 0.2, figsize = (20, 10)) # stacked = False : 다 겹치게 그래프 그려줌

plt.title("서울 -> 타시도 인구 이동", size = 30)
plt.ylabel("이동 인구 수", size = 20)
plt.xlabel("기간", size = 20)
plt.legend(fontsize = 15)

plt.show()
```

```python
# 면적 그래프 그리기
df_4.plot(kind = "area", stacked = True, alpha = 0.2, figsize = (20, 10)) # stacked = True : 데이터의 총 합 볼 수 있음, 데이터를 쌓아올림

plt.title("서울 -> 타시도 인구 이동", size = 30)
plt.ylabel("이동 인구 수", size = 20)
plt.xlabel("기간", size = 20)
plt.legend(fontsize = 15)

plt.show()
```

## 막대 그래프(bar plot)

- 데이터 값의 크기에 비례하여 높이를 갖는 직사각형 막대로 표현
- 막대 높이의 상대적 길이 차이를 통해 값의 크고 작음을 설명
- 세로형과 가로형 막대 그래프 두 종류가 있음
    - 다만 세로형은 선 그래프와 정보 제공 측면에서 큰 차이가 없음

```python
col_years = list(map(str, range(2010, 2018)))
df_4 = df_seoul.loc[["충청남도", "경상북도", "강원도", "전라남도"], col_years]
df_4 = df_4.transpose()
df_4.head()
```

```python
# 막대 그래프 그리기
df_4.plot(kind = "bar", figsize = (20, 10), width = 0.7,
         color = ["orange", "yellow", "skyblue", "pink"])

plt.title("서울 -> 타시도 인구 이동", size = 30)
plt.ylabel("이동 인구 수", size = 20)
plt.xlabel("기간", size = 20)
plt.legend(fontsize = 15)
plt.ylim(5000, 30000) # y축 범위 조정

plt.show()
```

```python
df_4 = df_4.transpose()
df_4.head()
```

```python
# 2010 - 2017년 이동 인구 수를 합계하여 새로운 열로 추가
df_4["합계"] = df_4.sum(axis = 1)
```

```python
df_4.head()
```

```python
# 가장 큰 값부터 정렬
df_total = df_4[["합계"]].sort_values(by = "합계", ascending = True) 
# ascending = True : 수가 높은 애부터 위에서 아래로 보여주고 싶으면 True 
df_total
```

```python
# 수평 막대 그래프 그리기
df_total.plot(kind = "barh", color = "cornflowerblue", width = 0.5, figsize = (10, 5))

plt.title("서울 -> 타시도 인구 이동")
plt.ylabel("전입지")
plt.xlabel("이동 인구 수")

plt.show()
```

### 보조 축 활용

- 보조 축을 추가하여 2개의 y축을 갖는 그래프 그리기

```python
df = pd.read_excel(r"C:\Users\USER\Desktop\project\pandas\data\남북한발전전력량.xlsx")
df.head()
```

```python
df = df.loc[5:9]
df.drop("전력량 (억㎾h)", axis = 1, inplace = True)
```

```python
df.set_index("발전 전력별", inplace = True)
```

```python
df = df.T
```

```python
df = df.rename(columns = {"합계" : "총발전량"})
```

```python
df.head()
```

```python
df["직전해 총발전량"] = df["총발전량"].shift(1)   # shift(1) : 한칸씩 미루는거임, 그 직전 해의 발전량
```

```python
# 증감율(변동률) 계산
df["증감율"] = ((df["총발전량"] / df["직전해 총발전량"]) -1) * 100
```

```python
df["증감율"]
```

```python
df.head()
```

```python
plt.rcParams["axes.unicode_minus"] = False   # 그래프 - 꺠지는 현상 방지
```

```python
# 2축 그래프 그리기
ax1 = df[["수력", "화력"]].plot(kind = "bar", figsize = (20, 10), width = 0.7, stacked = True)
ax2 = ax1.twinx()   # 원래 주어진 axis, x를 공유하는 쌍둥이 axis
ax2.plot(df.index, df["증감율"], ls = "--", marker = "o", markersize = 20, color = "yellow",
        label = "전년대비 증감율(%)")

ax1.set_ylim(0, 500)
ax2.set_ylim(-50, 50)

ax1.set_xlabel("연도", size = 20)       # x축 공유하는거라 ax1 으로 하건 ax2로 하건 노상관
ax1.set_ylabel("발전량")
ax2.set_ylabel("전년 대기 증감율(%)")

plt.title("북한 전력 발전량 (1990 ~ 2016)", size = 30)
ax1.legend(loc = "upper left")

plt.show()
```

## 산점도(scatter plot)

- 서로 다른 두 변수 사이의 관계를 나타냄
    - 2개의 연속 변수를 각각 x축과 y축에 하나씩 놓고 데이터 값이 위치하는 좌표를 찾아 점으로 표시
- 두 연속 변수의 관계를 보여주는 점에서 선 그래프와 유사
    - 선그래프를 그릴 때 "o" 옵션 등으로 선 없이 점으로만 표시하면 사실상 산점도라고 볼 수 있음

```python
# 스타일 초기화
plt.style.use("default")
# 하면 폰트 다시 실행
```

```python
# 연비(mpg)와 차중(weight) 열에 대한 산점도
df.plot(kind = "scatter", x = "weight", y = "mpg", c = "coral", s = 10, figsize = (10, 5)) # s = 10 : 사이즈
plt.title("산점도 mpg vs weight")
plt.show()
```

- cylinders 변수를 점의 크기로 표현하여 추가
    - 데이터 값을 그래도 쓰면 점의 크기 변화량이 너무 크기 때문에 0 ~ 1 범위의 실수 값으로 정규화

```python
# cylinders의 상대적 비율을 계산하여 시리즈 생성
cylinders_size = df["cylinders"] / df["cylinders"].max() * 200
```

```python
cylinders_size.tail()
```

```python
# 3개의 변수로 산점도 그리기
df.plot(kind = "scatter", x = "weight", y = "mpg", c = "coral", s = cylinders_size,
       figsize = (10, 5), alpha = 0.3)
plt.title("산점도 mpg, weight, cylinders")
plt.show()
```

### 그래프를 그림 파일로 저장

```python
cylinders_size /= 200

# 3개의 변수로 산점도 그리기
df.plot(kind = "scatter", x = "weight", y = "mpg", c = cylinders_size, s = 50,
       figsize = (10, 5), alpha = 0.3, marker = "+", cmap = "viridis")
plt.title("산점도 mpg, weight, cylinders")

plt.savefig("./scatter.png")        # 그림파일로 저장
plt.savefig("./scatter_transparent.png", transparent = True)   # 그림파일로 저장

plt.show()
```

## 파이 차트(pie chart)

- 원을 파이 조각처럼 나누어서 표현
- 조각의 크기는 해당 변수에 속하는 데이터 값의 크기에 비례

```python
# origin을 기준으로 그룹화
df_origin = df.groupby("origin").size()
df_origin
```

```python
# 제조국가 값을 실제 지역명으로 변경
df_origin.index = ["USA", "EU", "JPN"]
df_origin
```

```python
df_origin.plot(kind = "pie", figsize = (7, 5), autopct = "%.1f%%", startangle = 10,   # autopct = "%.1f%%" : 소숫점 첫째자리까지
              colors = ["chocolate", "bisque", "cadetblue"])
plt.title("차량 제조국가", size = 20)
plt.axis("equal") # 파이 차트의 비율을 동일하게 조정
plt.legend(labels = df_origin.index, loc = "upper right")    # loc = "upper right" : 범례를 우측 상단에 표시
plt.show()
```

## box plot

- 범주형 데이터의 분포를 파악하기에 적합

```python
# 그래프 객체 생성(figure에 2개의 서브플롯을 생성)
fig = plt.figure(figsize = (15, 5))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

# axe 객체에 boxplot 메서드로 그래프 출력
ax1.boxplot(x = [df[df["origin"] == 1]["mpg"],
                df[df["origin"] == 2]["mpg"],
                df[df["origin"] == 3]["mpg"]],
           labels = ["USA", "EU", "JAPAN"])

ax2.boxplot(x = [df[df["origin"] == 1]["mpg"],
                df[df["origin"] == 2]["mpg"],
                df[df["origin"] == 3]["mpg"]],
           labels = ["USA", "EU", "JAPAN"],
           vert = False)                           # vert = False : 수평박스플롯, True 라면 수직(기본값)

ax1.set_title("제조국가별 연비 분포(수직 박스 플롯)")
ax2.set_title("제조국가별 연비 분포(수평 박스 플롯)")
plt.show()
```
