# 문제3. 문자열 데이터를 입력 받고 capitalize() 메소드를 적용하는 capitalizer 함수 정의

```python
def capitalizer(input_str):              
    return input_str.capitalize()
```

```python
capitalizer("abc")
```

# 문제4. Mjob과 Fjob열에 capitalizer 함수 적용된 결과 출력

```python
# 방법 1

df["Mjob"]= df["Mjob"].map(capitalizer)
df["Fjob"] = df["Fjob"].map(capitalizer)
df
```

```python
# 방법2
df[["Mjob", "Fjob"]] = df[["Mjob", "Fjob"]].applymap(capitalizer)
df
```

- map : 시리즈의 객체 하나하나에 적용가능
- applymap : 데이터프레임의 객체 하나하나에 적용가능
- apply : 시리즈든 데이터프레임이든 완전 통째로만 적용가능

# 문제5. age가 17을 초과하는지 여부를 True, False로 표현하는 legal_drinker 열 생성

```python
df["legal_drinker"] = df["age"] > 17
df
```

```python
df["legal_drinker"] = df["age"].map(lambda x: x > 17)
# df["age"].map(lambda x: True if x > 17 else False)
df
```

```python
# 함수로 짜면
def is_legal(x):
#     return x > 17
    if x > 17:
        return True
    else :
        return False
```
