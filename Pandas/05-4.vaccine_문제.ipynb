# 문제4. day1num과 day2num의 월별 평균을 출력하기

```python
df.head()
```

```python
df.dtypes
```

```python
df["month"] = df["date"].str.split(".").str.get(1)   # 월에 대한 정보만 추출
```

```python
df.head()
```

```python
df.groupby("month")[["day1num", "day2num"]].mean()
```

- 아래 방법은 # datetime 활용

```python
df.head()
```

```python
# datetime 활용
df["date"] = pd.to_datetime(df["date"], format = "%Y.%m.%d")
```

```python
df.dtypes
```

```python
df["month"] = df["date"].map(lambda x: x.month)
```

```python
df.head()
```

```python
df.groupby("month")[["day1num", "day2num"]].mean()
```
